```python
import numpy as np
from PIL import Image

def calculate_psnr(img1, img2):
    if img1.shape != img2.shape:
        raise ValueError("Images must have the same shape")

    img1 = img1.astype(np.float32) / 255.0
    img2 = img2.astype(np.float32) / 255.0

    mse = np.mean((img1 - img2) ** 2)

    if mse < 1e-10:
        return float('inf')

    psnr = 20 * np.log10(1.0 / np.sqrt(mse))
    return psnr


img1 = Image.open("/root/code/OptiSplat/build/output/output_raw.jpg").convert("RGB")
img2 = Image.open("/root/code/OptiSplat/build/output/output.jpg").convert("RGB")
# img2 = Image.open("/root/code/FlashGS/test_out/_DSC8679.ppm").convert("RGB")

img1_np = np.array(img1)
img2_np = np.array(img2)

print("Shape1:", img1_np.shape)
print("Shape2:", img2_np.shape)

psnr = calculate_psnr(img1_np, img2_np)
print(f"PSNR: {psnr:.4f} dB")
img2_np = np.array(img2)

print(img1.mode)   # 应该是 'RGB'
print(img1.size)   # (width, height)

psnr = calculate_psnr(img1_np, img2_np)
print(f"PSNR: {psnr} dB")
```
